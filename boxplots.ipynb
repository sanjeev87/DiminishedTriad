```python
#this file makes box plots from logs generated by the hashing simulator
%matplotlib inline
import pylab
from pylab import *
from numpy import genfromtxt
import numpy as np

my_data = genfromtxt('hash_sim_log.txt', delimiter=',')
num_keys = np.unique(my_data[:,0])[~np.isnan(np.unique(my_data[:,0]))]
num_nodes = np.unique(my_data[:,1])[~np.isnan(np.unique(my_data[:,1]))]

for nodes in num_nodes:
    plotdata = []
    linesdata_x = []
    linesdata_y = []
    count = 1
    for key in num_keys:
        plotdata.append(my_data[(my_data[:, 0] == key) * (my_data[:, 1] == nodes)][:,3])
        linesdata_x.append(count)
        linesdata_y.append(key/nodes)
        count = count + 1
    pylab.boxplot(plotdata)
    pylab.xticks(range(1,len(num_keys)+1),num_keys/1000)
    pylab.title("NumNodes:["+str(nodes)+"]")
    pylab.xlabel('Num Keys Written (thousands)')
    pylab.ylabel('Num Keys Migrated')
    line1 = pylab.plot(linesdata_x,linesdata_y,c='black',linewidth=2)
    show()
    plotdata = []
```
